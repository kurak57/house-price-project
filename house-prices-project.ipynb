# Preprocessing Data

## Menangani Missing Value

```python
# memeriksa jumlah missing value
missing_values = train.isnull().sum()
missing_values[missing_values > 0]
```

```python
less = missing_values[missing_values < 1000].index
over = missing_values[missing_values > 1000].index
```

```python
# mengisi nilai hilang dengan median (kolom numerik)
numeric_columns = train[less].select_dtypes(include=['number']).columns
train[numeric_columns] = train[numeric_columns].fillna(train[numeric_columns].median())
train[numeric_columns]
```

```python
# mengisi nilai hilang pada kolom kategorikal
categorical_columns = train[less].select_dtypes(include='object').columns
for column in categorical_columns:
    train[column] = train[column].fillna(train[column].mode()[0])
```

```python
# menghapus kolom yang missiung valuenya > 75%
df = train.drop(columns=over)
```

```python
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

## Mengatasi Outlier

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# for column in numeric_columns:
#     plt.figure(figsize=(8,5))
#     sns.boxplot(x=df[column])
#     plt.title(f'Box Plot of {column}')
#     plt.show()
```

```python
# Menghitung IQR
Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# Filter dataframe untuk hanya menyimpan baris yang tidak mengandung outliers pada kolom numerik
condition = ~((df[numeric_columns] < lower_bound) | (df[numeric_columns] > upper_bound)).any(axis=1)
df_filtered_numeric = df.loc[condition, numeric_columns]

categorical_columns = df.select_dtypes(include='object').columns
df = pd.concat([df_filtered_numeric, df.loc[condition, categorical_columns]], axis=1)
df
```

## Persebaran Data Histogram Tiap Fitur

```python
# menghitung jumlah variable
num_vars = df_lencoder.shape[1]

#menentukan jumlah baris dan kolom untuk grid subplot
n_cols = 4 #jumlah kolom plot yang diinginkan
n_rows = -(-num_vars // n_cols)

#membuat subplot
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows*4))

axes = axes.flatten()

for i, column in enumerate(df_lencoder.columns):
    df_lencoder[column].hist(ax=axes[i], bins=20, edgecolor='black')
    axes[i].set_title(column)
    axes[i].set_xlabel('Value')
    axes[i].set_ylabel('Frequency')
    
#mengahpu axex/ plot kosong
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])
    
plt.tight_layout()
plt.show()
```

## Korelasi dan distribusi beberapa kolom

```python
import matplotlib.pyplot as plt
import seaborn as sns

# memilih beberapa kolom
column_to_plot = ['OverallQual', 'YearBuilt', 'LotArea', 'SaleType', 'SaleCondition']

plt.figure(figsize=(15, 10))

for i, column in enumerate(column_to_plot,1):
    plt.subplot(2,3,i)
    sns.histplot(df_lencoder[column], kde=True, bins= 30)
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```

```python
# Visualisasi korelasi antar variabel numerik
plt.figure(figsize=(12,10))
corelation_matrix = df_lencoder.corr()

sns.heatmap(corelation_matrix, annot=False, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Correlation Matrix')
plt.show()
```

```python
# melihat korelasi sale price (sebagai varibel target) dengan variable lain
target_corr = df_lencoder.corr()['SalePrice']

target_corr_sorted = target_corr.abs().sort_values(ascending=False)

plt.figure(figsize=(10,6))
target_corr_sorted.plot(kind='bar')
plt.title('Correlatio with SalePrice')
plt.ylabel('Correlation Coefficient')
plt.xlabel('Variables')

plt.show()
```
