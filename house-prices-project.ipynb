# Preprocessing Data

## Menangani Missing Value

```python
# memeriksa jumlah missing value
missing_values = train.isnull().sum()
missing_values[missing_values > 0]
```

```python
less = missing_values[missing_values < 1000].index
over = missing_values[missing_values > 1000].index
```

```python
# mengisi nilai hilang dengan median (kolom numerik)
numeric_columns = train[less].select_dtypes(include=['number']).columns
train[numeric_columns] = train[numeric_columns].fillna(train[numeric_columns].median())
train[numeric_columns]
```

```python
# mengisi nilai hilang pada kolom kategorikal
categorical_columns = train[less].select_dtypes(include='object').columns
for column in categorical_columns:
    train[column] = train[column].fillna(train[column].mode()[0])
```

```python
# menghapus kolom yang missiung valuenya > 75%
df = train.drop(columns=over)
```

```python
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

## Mengatasi Outlier

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# menghitung jumlah variable
num_vars = df.shape[1]

#menentukan jumlah baris dan kolom untuk grid subplot
n_cols = 4 #jumlah kolom plot yang diinginkan
n_rows = -(-num_vars // n_cols)

#membuat subplot
fig, axes = plt.subplots(n_rows, n_cols, figsize=(20, n_rows*4))

axes = axes.flatten()

for i, column in enumerate(numeric_columns):
    sns.boxplot(x=df[column], ax=axes[i])
    axes[i].set_title(f'Box Plot of {column}')
    
#mengahpu axex/ plot kosong
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])
    
plt.tight_layout()
plt.show()
```

```python
# Menghitung IQR
Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
# Filter dataframe untuk hanya menyimpan baris yang tidak mengandung outliers pada kolom numerik
condition = ~((df[numeric_columns] < lower_bound) | (df[numeric_columns] > upper_bound)).any(axis=1)
df_filtered_numeric = df.loc[condition, numeric_columns]

categorical_columns = df.select_dtypes(include='object').columns
df = pd.concat([df_filtered_numeric, df.loc[condition, categorical_columns]], axis=1)
df
```

# Data Splitting

```python
import sklearn
```

```python
# Memisah fitur (X) dan target (y)
x = df_lencoder.drop(columns=['SalePrice'])  
y = df_lencoder['SalePrice'] 
```

```python
# membagi dataset menjadi train dan test
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1)
```

```python
# jumlah data fitur

print("Jumlah data:", len(x))
print("Jumlah data latih:", len(x_train))
print("Jumlah data test:", len(x_test))
```

# Melatih Model

```python
# least angle regression
from sklearn import linear_model
lars = linear_model.Lars(n_nonzero_coefs=1).fit(x_train, y_train)

# Linear Regression
from sklearn.linear_model import LinearRegression
LR = LinearRegression().fit(x_train,y_train)

#Gradient Boosting Regressor
from sklearn.ensemble import GradientBoostingRegressor
GBR = GradientBoostingRegressor(random_state=184)
GBR.fit(x_train, y_train)
```

## Evaluasi LARS model

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

pred_lars = lars.predict(x_test)
mae_lars = mean_absolute_error(y_test, pred_lars)
mse_lars = mean_squared_error(y_test, pred_lars)
r2_lars = r2_score(y_test, pred_lars)
```

```python
data = {
    'MAE': [mae_lars],
    'MSE': [mse_lars],
    'R2': [r2_lars]
}
 
df_results = pd.DataFrame(data, index=['Lars'])
df_results
```

## Evaluasi LR mmodel

```python
pred_lr = LR.predict(x_test)
mae_lr = mean_absolute_error(y_test, pred_lr)
mse_lr = mean_squared_error(y_test, pred_lr)
r2_lr = r2_score(y_test, pred_lr)

df_results.loc["Linear Regression"] = [mae_lr, mse_lr, r2_lr]
df_results
```

## Evaluasi GBR

```python
pred_gbr = GBR.predict(x_test)
mae_gbr = mean_absolute_error(y_test,pred_gbr)
mse_gbr = mean_squared_error(y_test, pred_gbr)
r2_gbr = r2_score(y_test, pred_gbr)

df_results.loc['Gradient Boost Regressor'] = [mae_gbr, mse_gbr, r2_gbr]
df_results
```

## Flask API

```python
import threading
from flask import Flask, request, jsonify

import joblib

app = Flask(__name__)

# load model
joblib_model = joblib.load('gbr_model.joblib') 

# Route untuk halaman home ("/")
@app.route('/')
def home():
    return "Welcome to the Machine Learning Prediction API!"

# Route untuk prediksi ("/predict")
@app.route('/predict', methods=['POST'])
def predict():
    data = request.json.get("data")  # Mengambil data dari request JSON
    if not data:
        return jsonify({"error": "No data provided"}), 400
    prediction = joblib_model.predict(data)
    return jsonify({'prediction': prediction.tolist()})

# Fungsi untuk menjalankan Flask di thread terpisah
def run_flask():
    app.run(debug=True, use_reloader=False, host='0.0.0.0', port=5000)

# Menjalankan Flask di thread terpisah
flask_thread = threading.Thread(target=run_flask)
flask_thread.start()
```
